# C1. A horizontal bar plot that shows the top 10 populated communes. 

```python
import pandas as pd
import numpy as np
df = pd.read_csv("p1_communes.csv") #Importing the data as a Pandas DataFrame, named as df
df.dropna(inplace=True) #removing missing data from the dataset
df.set_index("Commune", inplace=True)

# Add titles, labels, invert y-axis

Populated_comm = df["Residents"].sort_values (ascending=False).head(10)
colormap = plt.cm.tab10(range(0, len(Populated_comm)))
bar_plot = Populated_comm.plot.barh(color=colormap)
bar_plot.set_title("The 10 most populated communes")
bar_plot.set_xlabel("Residents")
bar_plot.invert_yaxis()
plt.show(bar_plot)
```

# C2. To plot a horizontal stacked bar chart that shows how the lands of the 10 most populated communes are divided into the 4 area types: Settlement, Agricultural, Wooded, Unproductive. 

```python
columns = [
    "Settlement area",
    "Agricultural area",
    "Wooded area",
    "Unproductive area",
    "Commune"
]
        
# Load the DataFrame, we will work on the selected columns
Area_type = pd.read_csv("p1_communes.csv", usecols=columns)
Commune_index = Area_type.set_index("Commune")
Selected_communes = Commune_index.loc[["Zürich", "Genève", "Basel", "Lausanne", "Bern", "Winterthur", "Luzern", "St. Gallen", "Lugano", "Biel/Bienne"]]
Selected_communes.plot.barh(stacked=True)
plt.legend(loc='lower left', bbox_to_anchor=(1, 0.5))
plt.xlabel("Area type in %")
plt.title("10 populated communes by type of area")
plt.show()
```

# D. Probabilities

# Compute the probability that a randomly selected commune with elevation over 2000 is from the canton of Valais.

```python
df = pd.read_csv("p1_communes.csv") #Importing the data as a Pandas DataFrame, named as df
df.dropna(inplace=True) #removing missing data from the dataset
df.set_index("Commune", inplace=True)

df['High_commune'] = np.where(df['Elevation'] > 2000, 1, 0)
df['Valais'] = np.where(df['Canton'] == 'VS', 1, 0)
df['count'] = 1
df = df[['High_commune','Valais','count']]
df
```

```python
pd.pivot_table(
    df, 
    values='count', 
    index=['Valais'], 
    columns=['High_commune'], 
    aggfunc=np.size, 
    fill_value=0
)
```
